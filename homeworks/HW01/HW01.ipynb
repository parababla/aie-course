### Примеры индексирования и срезов

Пример на одномерном массиве. Получение отдельного элемента и срезы

```python
example_array = np.arange(10)
print("Элемент №3:", example_array[3])
print("Элемент №3 и №5:", example_array[[3, 5]])
print("Срез от 2-ого элемента до 5-ого (вкл):", example_array[2:6])
print("Обратный порядок:", example_array[::-1]) # начало и конец по умолчаению, шаг -1
```

Пример на многомерном массиве. Получение отдельного элемента и срезы

```python
example_matrix = np.arange(32)
example_matrix.shape = (2, 4, 4)
print(example_matrix)
```

Первый индекс — номер матрицы;

второй индекс — номер строки;

третий индекс — номер столбца;

```python
print("Элемент №15:", example_matrix[(0, 3, 3)])
print("Элемент №27:", example_matrix[1][2][3])
print("Элемент №6:", example_matrix[0, 1, 2])
print("2-ая строка 2-ой матрицы:", example_matrix[1, 1:2])
print("Последний столбец обеих матриц:\n", example_matrix[:,:,-1])
```

### Примеры векторных операций и базовых статистик

Создадим два двумерных массива для тестирования операций

```python
first_number_array = np.arange(10)
first_number_array.shape = (2, 5)
second_number_array = np.arange(10, 20)
second_number_array.shape = (2, 5)
print(first_number_array, '\n\n', second_number_array)
```

```python
addition = first_number_array + second_number_array
print("Сложение:\n", addition)
subtraction = first_number_array - second_number_array
print("Вычитание:\n", subtraction)
multiplication = first_number_array * second_number_array
print("Умножение:\n", multiplication)
division = second_number_array / first_number_array
print("Деление:\n", division)
```

Вычисление простой статистики

```python
print("Сумма значений первой матрицы:", np.sum(first_number_array))
print("Минимальное значение первой матрицы:", np.min(first_number_array))
print("Максимальное значение первой матрицы:", np.max(first_number_array))
print("Среднее значение первой матрицы:", np.mean(first_number_array))
print("Медиана первой матрицы:", np.median(first_number_array))
print("Стандартное отклонение первой матрицы:", np.std(first_number_array))
print("Дисперсия первой матрицы:", np.var(first_number_array))
```
